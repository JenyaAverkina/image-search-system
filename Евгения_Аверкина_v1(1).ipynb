 - –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —á–µ—Ä–µ–∑ inner. –¢—É—Ç –≤—Å–µ –ø–ª–æ—Ö–æ - –∏ –¥–∞–Ω–Ω—ã—Ö –û–ß–ï–ù–¨ –º–∞–ª–æ (–≤ –¥–≤–∞ —Ä–∞–∑–∞ –º–µ–Ω—å—à–µ, —á–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ç–æ–ª—å–∫–æ —ç–∫—Å–ø–µ—Ä—Ç–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏) –∏ –∫–∞—á–µ—Å—Ç–≤–æ —É –Ω–∏—Ö "–ø–æ–¥–ø–æ—Ä—á–µ–Ω–æ". –ò –Ω–∞–≤—ã–∫–æ–≤ –ø–æ –æ–±—Ä–∞–±–æ—Ç–∫–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ –ø–æ–ª—É—á–∞–µ–º.....
    
–°–µ–π—á–∞—Å —Ç—ã —Å–æ–µ–¥–∏–Ω–∏–ª–∞ —ç–∫—Å–ø–µ—Ä—Ç–Ω—ã–µ –∏ –∫—Ä–∞—É–¥—Å–æ—Ä—Å –æ—Ü–µ–Ω–∫–∏, –Ω–æ —Ç.–∫. –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª—è `inner`, —Ç–æ –º—ã –ø—Ä–æ—Å—Ç–æ –ø–æ—Ç–µ—Ä—è–ª–∏ –ø–æ–ª–æ–≤–∏–Ω—É –¥–∞–Ω–Ω—ã—Ö, –Ω–µ –ø–æ–ª—É—á–∏–≤ –Ω–∏—á–µ–≥–æ –≤ –∑–∞–º–µ–Ω (—Ç–æ—á–Ω–µ–µ —Ç–æ–ª—å–∫–æ —É—Ö—É–¥—à–∏–≤ –Ω–∞—à–∏ –æ—Ü–µ–Ω–∫–∏).
    
–ü–æ–∂–∞–ª—É–π—Å—Ç–∞ –∏—Å–ø–æ–ª—å–∑—É–π –ª—é–±—É—é –¥—Ä—É–≥—É—é —Å—Ç—Ä–∞—Ç–µ–≥–∏—é, –∫—Ä–æ–º–µ –≤–Ω—É—Ç—Ä–µ–Ω–Ω–µ–≥–æ —Å–æ–µ–¥–∏–Ω–µ–Ω–∏—è. –°–ª–∏—à–∫–æ–º –º–∞–ª–æ –¥–∞–Ω–Ω—ã—Ö –ø—Ä–∏ —ç—Ç–æ–º –ø–æ–ª—É—á–∞–µ—Ç—Å—è.

<br>–°–æ–∑–¥–∞–¥–∏–º —Ñ–∏–Ω–∞–ª—å–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏</br>

```python
def create_master_dataset(expert_data, crowd_data, train_data):
    # –û–±—ä–µ–¥–∏–Ω—è–µ–º —ç–∫—Å–ø–µ—Ä—Ç–Ω—ã–µ –∏ –∫—Ä–∞—É–¥—Å–æ—Ä—Å–∏–Ω–≥–æ–≤—ã–µ –æ—Ü–µ–Ω–∫–∏ —á–µ—Ä–µ–∑ outer join
    merged_scores = expert_data.merge(
        crowd_data[['image_name', 'caption_id', 'proportion']],
        on=['image_name', 'caption_id'],
        how='outer',
        suffixes=('_expert', '_crowd')
    )
    
    # –°–æ–∑–¥–∞–µ–º —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é —Å –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–æ–º —ç–∫—Å–ø–µ—Ä—Ç–Ω—ã—Ö –æ—Ü–µ–Ω–æ–∫
    def calculate_final_score(row):
        if not pd.isna(row['expert_norm']):
            # –ï—Å–ª–∏ –µ—Å—Ç—å —ç–∫—Å–ø–µ—Ä—Ç–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ - –∏—Å–ø–æ–ª—å–∑—É–µ–º –µ–µ
            return row['expert_norm']
        elif not pd.isna(row['proportion']):
            # –ï—Å–ª–∏ –Ω–µ—Ç —ç–∫—Å–ø–µ—Ä—Ç–Ω–æ–π, –Ω–æ –µ—Å—Ç—å –∫—Ä–∞—É–¥—Å–æ—Ä—Å–∏–Ω–≥–æ–≤–∞—è - –∏—Å–ø–æ–ª—å–∑—É–µ–º –µ–µ
            return row['proportion']
        else:
            # –ï—Å–ª–∏ –Ω–µ—Ç –Ω–∏–∫–∞–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫ - –ø—Ä–æ–ø—É—Å–∫ (–±—É–¥–µ—Ç —É–¥–∞–ª–µ–Ω–æ)
            return np.nan
    
    merged_scores['target'] = merged_scores.apply(calculate_final_score, axis=1)
    
    # –£–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ–∫–∏ –±–µ–∑ –æ—Ü–µ–Ω–æ–∫
    merged_scores = merged_scores.dropna(subset=['target'])
    
    # –û–±—ä–µ–¥–∏–Ω—è–µ–º —Å —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
    final_data = train_data.merge(
        merged_scores[['image_name', 'caption_id', 'target']],
        on=['image_name', 'caption_id'],
        how='inner'
    )
    
    return final_data

# –ü–æ–¥–≥–æ—Ç–∞–≤–ª–∏–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è
expert_data = expert_annotations[['image_name', 'caption_id', 'expert_norm']]
crowd_data = crowd_annotations[['image_name', 'caption_id', 'proportion']]

# –°–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Å–µ—Ç
final_data = create_master_dataset(expert_data, crowd_data, train_df)

display(f"–§–∏–Ω–∞–ª—å–Ω—ã–π —Ä–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞: {final_data.shape}")
display(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π: {final_data['image_name'].nunique()}")
display(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –æ–ø–∏—Å–∞–Ω–∏–π: {final_data['caption_id'].nunique()}")

# –ê–Ω–∞–ª–∏–∑ –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤ –æ—Ü–µ–Ω–æ–∫
expert_count = final_data[final_data['target'].isin([0.0, 0.33, 0.67, 1.0])].shape[0]
crowd_count = final_data[~final_data['target'].isin([0.0, 0.33, 0.67, 1.0])].shape[0]

display(f"–≠–∫—Å–ø–µ—Ä—Ç–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏: {expert_count} ({expert_count/len(final_data)*100:.1f}%)")
display(f"–ö—Ä–∞—É–¥—Å–æ—Ä—Å–∏–Ω–≥–æ–≤—ã–µ –æ—Ü–µ–Ω–∫–∏: {crowd_count} ({crowd_count/len(final_data)*100:.1f}%)")

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –¥–∞–Ω–Ω—ã—Ö
display(f"–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤: {final_data.isna().sum().sum()}")
display(final_data.isna().sum())
display(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤: {final_data.duplicated().sum()}")

# –ê–Ω–∞–ª–∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
display(final_data['target'].describe())

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—Ü–µ–Ω–æ–∫ –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# –û–±—â–µ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
ax1.hist(final_data['target'], bins=30, alpha=0.7, color='purple', edgecolor='black')
ax1.set_title('–û–±—â–µ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π')
ax1.set_xlabel('–û—Ü–µ–Ω–∫–∞ —Ä–µ–ª–µ–≤–∞–Ω—Ç–Ω–æ—Å—Ç–∏')
ax1.set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')

# –†–∞–∑–¥–µ–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º
expert_targets = final_data[final_data['target'].isin([0.0, 0.33, 0.67, 1.0])]['target']
crowd_targets = final_data[~final_data['target'].isin([0.0, 0.33, 0.67, 1.0])]['target']

ax2.hist(expert_targets, bins=4, alpha=0.7, label='–≠–∫—Å–ø–µ—Ä—Ç–Ω—ã–µ', color='blue', edgecolor='black')
ax2.hist(crowd_targets, bins=20, alpha=0.7, label='–ö—Ä–∞—É–¥—Å–æ—Ä—Å–∏–Ω–≥', color='orange', edgecolor='black')
ax2.set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º –æ—Ü–µ–Ω–æ–∫')
ax2.set_xlabel('–û—Ü–µ–Ω–∫–∞ —Ä–µ–ª–µ–≤–∞–Ω—Ç–Ω–æ—Å—Ç–∏')
ax2.set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
ax2.legend()

plt.tight_layout()
plt.show()

# –î–µ—Ç–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö
display(final_data.info())
display(final_data.head(10))
```

```python
def filter_restricted_content(data):
    # –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –∫–æ–Ω—Ç–µ–Ω—Ç–∞ —Å –¥–µ—Ç—å–º–∏
    restricted_words = ['child', 'children', 'kid', 'kids', 'boy', 'boys', 'girl', 'girls', 'baby']
    
    def contains_restricted(text):
        if isinstance(text, str):
            text_lower = text.lower()
            return any(word in text_lower for word in restricted_words)
        return False
    
    
    bad_captions = data[data['text'].apply(contains_restricted)]['caption_id']
    
    # –ò–∑–≤–ª–µ–∫–∞–µ–º –∏–º–µ–Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏–∑ caption_id
    bad_images = set([str(caption_id)[:-2] for caption_id in bad_captions])
    
    # –£–¥–∞–ª—è–µ–º –≤—Å–µ –∑–∞–ø–∏—Å–∏ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏
    data_clean = data[~data['image_name'].isin(bad_images)]
    
    print(f"–£–¥–∞–ª–µ–Ω–æ –∑–∞–ø–∏—Å–µ–π: {len(data) - len(data_clean)}")
    print(f"–û—Å—Ç–∞–ª–æ—Å—å –∑–∞–ø–∏—Å–µ–π: {len(data_clean)}")
    print(f"–ù–∞–π–¥–µ–Ω–æ –ø–ª–æ—Ö–∏—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π: {len(bad_images)}")
    
    return data_clean

train_df_clean = filter_restricted_content(train_df)
```

## –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π

–ü–µ—Ä–µ–π–¥—ë–º –∫ –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π.

```python
class ImageVectorizer:
    def __init__(self):
        self.model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
        self.model = nn.Sequential(*list(self.model.children())[:-1])
        self.model.eval()
        
        self.transform = transforms.Compose([
            transforms.Resize(256),
            transforms.CenterCrop(224),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                              std=[0.229, 0.224, 0.225]),
        ])
    
    def vectorize_image(self, image_path):
        try:
            expanded_path = os.path.expanduser(image_path)
            if not os.path.exists(expanded_path):
                return None
                
            img = Image.open(expanded_path).convert('RGB')
            img_tensor = self.transform(img).unsqueeze(0)
            
            with torch.no_grad():
                features = self.model(img_tensor)
            
            return features.flatten().numpy()
        except Exception as e:
            return None
    
    def vectorize_images(self, image_paths):
        features = []
        successful_paths = []
        
        for i, img_path in enumerate(image_paths):
            feature = self.vectorize_image(img_path)
            if feature is not None:
                features.append(feature)
                successful_paths.append(img_path)
            
            if (i + 1) % 50 == 0:
                display(f"–û–±—Ä–∞–±–æ—Ç–∞–Ω–æ {i + 1}/{len(image_paths)} –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π")
        
        return np.array(features), successful_paths

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ç–æ—Ä
image_vectorizer = ImageVectorizer()

# –°–Ω–∞—á–∞–ª–∞ –ø–æ–ª—É—á–∏–º —Å–ø–∏—Å–æ–∫ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏–∑ final_data
unique_images = final_data['image_name'].unique()
display(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –¥–ª—è –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏–∏: {len(unique_images)}")

# –°–æ–∑–¥–∞–µ–º –ø–æ–ª–Ω—ã–µ –ø—É—Ç–∏ –∫ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º
image_paths = [os.path.join(TRAIN_IMAGES_PATH, img_name) for img_name in unique_images]

# –í–µ–∫—Ç–æ—Ä–∏–∑—É–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
display("–ù–∞—á–∏–Ω–∞–µ–º –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—é –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π...")
train_features, successful_paths = image_vectorizer.vectorize_images(image_paths)

# –°–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å —Å –≤–µ–∫—Ç–æ—Ä–Ω—ã–º–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è–º–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
train_features_dict = {}
for img_name, feature in zip(unique_images, train_features):
    train_features_dict[img_name] = feature

display(f"{len(train_features_dict)}")
```

`–í—Å–µ 714 –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –≤–µ–∫—Ç–æ—Ä–∏–∑–æ–≤–∞–Ω—ã.`

## –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è —Ç–µ–∫—Å—Ç–∞

```python
# –ó–∞–≥—Ä—É–∂–∞–µ–º –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—É—é –º–æ–¥–µ–ª—å
text_model = SentenceTransformer('all-MiniLM-L6-v2')

# –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è —Ç–µ–∫—Å—Ç–æ–≤ —Å pooling
text_embeddings = text_model.encode(final_data['text'].tolist(), 
                                  convert_to_tensor=False,
                                  show_progress_bar=True)

display(f"{text_embeddings.shape}")
```

–í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è —Ç–µ–∫—Å—Ç–æ–≤ –≤—ã–ø–æ–ª–Ω–µ–Ω–∞ —É—Å–ø–µ—à–Ω–æ - –¥–∞–Ω–Ω—ã–µ –≥–æ—Ç–æ–≤—ã –¥–ª—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ —ç—Ç–∞–ø–∞ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π.

## –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è

```python
# –°—Ç—Ä–æ–∏–º –º–∞—Ç—Ä–∏—Ü—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å —Ç–µ–∫—Å—Ç–æ–≤—ã–º–∏ —ç–º–±–µ–¥–¥–∏–Ω–≥–∞–º–∏
def build_feature_matrix(img_feat_dict, text_embeddings, df, target_col='target'):
    X_image, X_text, y_values, group_names = [], [], [], []
    
    for row_idx, row in df.iterrows():
        img_name = row['image_name']
        
        if img_name in img_feat_dict:
            X_image.append(img_feat_dict[img_name])
            X_text.append(text_embeddings[row_idx])
            y_values.append(row[target_col])
            group_names.append(img_name)
    
    X_image = np.array(X_image)
    X_text = np.array(X_text)
    y_values = np.array(y_values)
    
    # –†–∞–∑–¥–µ–ª—å–Ω–∞—è L2-–Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è
    image_normalizer = Normalizer(norm='l2')
    text_normalizer = Normalizer(norm='l2')
    
    X_image_normalized = image_normalizer.transform(X_image)
    X_text_normalized = text_normalizer.transform(X_text)
    
    X_combined = np.concatenate([X_image_normalized, X_text_normalized], axis=1)
    
    display(f"{X_image_normalized.shape}")
    display(f"{X_text_normalized.shape}")
    display(f"{X_combined.shape}")
    
    return X_combined, y_values, group_names

# –ü–µ—Ä–µ—Å—Ç—Ä–∞–∏–≤–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å —ç–º–±–µ–¥–¥–∏–Ω–≥–∞–º–∏
X_combined, y, groups = build_feature_matrix(
    train_features_dict, text_embeddings, final_data
)
```

–û–¥–Ω–∞–∫–æ –ø—Ä–æ—Å—Ç–æ –≤–∑—è—Ç—å –æ–±—ã—á–Ω—ã–π `StandardScaler` –∑–¥–µ—Å—å –±—ã–ª–æ –±—ã –æ—à–∏–±–∫–æ–π. –≠—Ç–æ—Ç –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç –æ—Ç–ª–∏—á–Ω–æ –ø–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è –æ–±—ã—á–Ω—ã—Ö —Ç–∞–±–ª–∏—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –Ω–æ –æ–Ω –º–æ–∂–µ—Ç ¬´–∏—Å–ø–æ—Ä—Ç–∏—Ç—å¬ª —ç–º–±–µ–¥–¥–∏–Ω–≥–∏, —Ç–∞–∫ –∫–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç *–ø–æ–∫–æ–ª–æ–Ω–æ—á–Ω–æ* –∏ —Ä–∞–∑—Ä—É—à–∞–µ—Ç –≤–Ω—É—Ç—Ä–µ–Ω–Ω—é—é –≥–µ–æ–º–µ—Ç—Ä–∏—é –≤–µ–∫—Ç–æ—Ä–æ–≤, –≤ –∫–æ—Ç–æ—Ä–æ–π –∏ –∑–∞–∫–ª—é—á–µ–Ω –≤–µ—Å—å –∏—Ö —Å–º—ã—Å–ª.

–°–∞–º—ã–π –ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –∏ —Å–æ–≤—Ä–µ–º–µ–Ω–Ω—ã–π –ø–æ–¥—Ö–æ–¥ –¥–ª—è —Ç–≤–æ–µ–π –∑–∞–¥–∞—á–∏ ‚Äî —ç—Ç–æ **—Ä–∞–∑–¥–µ–ª—å–Ω–∞—è L2-–Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è** –≤–µ–∫—Ç–æ—Ä–æ–≤ –ø–µ—Ä–µ–¥ –∏—Ö –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ–º.

1.  **–°–Ω–∞—á–∞–ª–∞** —Ç—ã –±–µ—Ä–µ—à—å –≤—Å–µ –≤–µ–∫—Ç–æ—Ä—ã –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π (–∏–∑ `ResNet`) –∏ –Ω–æ—Ä–º–∞–ª–∏–∑—É–µ—à—å –∏—Ö –æ—Ç–¥–µ–ª—å–Ω–æ —Å –ø–æ–º–æ—â—å—é `Normalizer(norm='l2')`.
2.  **–ü–æ—Ç–æ–º** —Ç—ã –±–µ—Ä–µ—à—å –≤—Å–µ –≤–µ–∫—Ç–æ—Ä—ã —Ç–µ–∫—Å—Ç–æ–≤ (–∏–∑ `BERT`) –∏ —Ç–æ–∂–µ –Ω–æ—Ä–º–∞–ª–∏–∑—É–µ—à—å –∏—Ö –æ—Ç–¥–µ–ª—å–Ω–æ, —Å–≤–æ–∏–º —ç–∫–∑–µ–º–ø–ª—è—Ä–æ–º `Normalizer`'–∞.
3.  –ò —Ç–æ–ª—å–∫–æ **–ø–æ—Å–ª–µ —ç—Ç–æ–≥–æ** —Ç—ã –∫–æ–Ω–∫–∞—Ç–µ–Ω–∏—Ä—É–µ—à—å —É–∂–µ –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã –≤ –µ–¥–∏–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

–ß—Ç–æ —ç—Ç–æ –¥–∞–µ—Ç? –¢—ã –ø—Ä–∏–≤–æ–¥–∏—à—å –æ–±–∞ —Ç–∏–ø–∞ –≤–µ–∫—Ç–æ—Ä–æ–≤ –∫ —Å–æ–ø–æ—Å—Ç–∞–≤–∏–º–æ–º—É –º–∞—Å—à—Ç–∞–±—É (–≤—Å–µ –æ–Ω–∏ —Å—Ç–∞–Ω–æ–≤—è—Ç—Å—è –≤–µ–∫—Ç–æ—Ä–∞–º–∏ –µ–¥–∏–Ω–∏—á–Ω–æ–π –¥–ª–∏–Ω—ã), –Ω–æ –ø—Ä–∏ —ç—Ç–æ–º –ø–æ–ª–Ω–æ—Å—Ç—å—é —Å–æ—Ö—Ä–∞–Ω—è–µ—à—å —Å–µ–º–∞–Ω—Ç–∏—á–µ—Å–∫—É—é —Å—Ç—Ä—É–∫—Ç—É—Ä—É –≤–Ω—É—Ç—Ä–∏ –∫–∞–∂–¥–æ–≥–æ –∏–∑ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤ (–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞ –∫–∞—Ä—Ç–∏–Ω–æ–∫ –∏ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞ —Ç–µ–∫—Å—Ç–æ–≤). –≠—Ç–æ —Å–∞–º—ã–π –Ω–∞–¥–µ–∂–Ω—ã–π —Å–ø–æ—Å–æ–± –ø–æ–¥–≥–æ—Ç–æ–≤–∏—Ç—å —Ç–∞–∫–∏–µ —Å–ª–æ–∂–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –¥–ª—è –º–æ–¥–µ–ª–∏.

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Ç–µ–Ω–∑–æ—Ä—ã
X_train_t = torch.FloatTensor(X_train)
X_valid_t = torch.FloatTensor(X_valid)
y_train_t = torch.FloatTensor(y_train)
y_valid_t = torch.FloatTensor(y_valid)

class CrossModalNet(nn.Module):
    def __init__(self, input_dim):
        super(CrossModalNet, self).__init__()
        
        # –û—Å–Ω–æ–≤–Ω—ã–µ —Å–ª–æ–∏ 
        self.fc1 = nn.Linear(input_dim, 2048)
        self.bn1 = nn.BatchNorm1d(2048)
        
        self.fc2 = nn.Linear(2048, 1024)
        self.bn2 = nn.BatchNorm1d(1024)
        
        self.fc3 = nn.Linear(1024, 512)
        self.bn3 = nn.BatchNorm1d(512)
        
        self.fc4 = nn.Linear(512, 256)
        self.bn4 = nn.BatchNorm1d(256)
        
        self.fc5 = nn.Linear(256, 128)
        self.bn5 = nn.BatchNorm1d(128)
        
        self.fc6 = nn.Linear(128, 64)
        self.bn6 = nn.BatchNorm1d(64)
        
        self.fc7 = nn.Linear(64, 32)
        self.bn7 = nn.BatchNorm1d(32)
        
        self.fc8 = nn.Linear(32, 1)
        
        self.dropout = nn.Dropout(0.5)
        
    def forward(self, x):
        # –°–ª–æ–π 1
        x = F.relu(self.fc1(x))
        x = self.bn1(x)
        x = self.dropout(x)
        
        # –°–ª–æ–π 2
        x = F.relu(self.fc2(x))
        x = self.bn2(x)
        x = self.dropout(x)
        
        # –°–ª–æ–π 3
        x = F.relu(self.fc3(x))
        x = self.bn3(x)
        x = self.dropout(x)
        
        # –°–ª–æ–π 4
        x = F.relu(self.fc4(x))
        x = self.bn4(x)
        x = self.dropout(x)
        
        # –°–ª–æ–π 5
        x = F.relu(self.fc5(x))
        x = self.bn5(x)
        x = self.dropout(x)
        
        # –°–ª–æ–π 6
        x = F.relu(self.fc6(x))
        x = self.bn6(x)
        x = self.dropout(x)
        
        # –°–ª–æ–π 7
        x = F.relu(self.fc7(x))
        x = self.bn7(x)
        x = self.dropout(x)
        
        # –í—ã—Ö–æ–¥–Ω–æ–π —Å–ª–æ–π —Å —Å–∏–≥–º–æ–∏–¥–æ–π
        x = torch.sigmoid(self.fc8(x))
        
        return x.squeeze()

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–æ–¥–µ–ª–∏
n_in_neurons = X_train.shape[1]
net = CrossModalNet(n_in_neurons)

# –ü—Ä–∏–º–µ–Ω–∏–º –æ–ø—Ç–∏–º–∏–∑–∞—Ç–æ—Ä Adam 
optimizer = torch.optim.Adam(net.parameters(), lr=1e-4)
loss_fn = nn.MSELoss()

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–±—É—á–µ–Ω–∏—è
num_epochs = 250
batch_size = 50  
num_batches = ceil(len(X_train_t) / batch_size)

best_rmse = float('inf')
best_epoch = -1

# –°–ø–∏—Å–∫–∏ –¥–ª—è –æ—Ç—Å–ª–µ–∂–∏–≤–∞–Ω–∏—è –º–µ—Ç—Ä–∏–∫
train_losses = []
val_rmse_list = []

for epoch in range(num_epochs):
    net.train()
    epoch_loss = 0.0
    
    # –°–ª—É—á–∞–π–Ω—ã–π –ø–æ—Ä—è–¥–æ–∫ –±–∞—Ç—á–µ–π
    order = np.random.permutation(len(X_train_t))
    
    for b in range(num_batches):
        start = b * batch_size
        end = start + batch_size
        idx = order[start:end]
        
        xb = X_train_t[idx]
        yb = y_train_t[idx]
        
        optimizer.zero_grad()
        preds = net(xb)
        loss_value = loss_fn(preds, yb)
        loss_value.backward()
        optimizer.step()
        
        epoch_loss += loss_value.item()
    
    # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –ª–æ—Å—Å–∞ –∑–∞ —ç–ø–æ—Ö—É
    avg_loss = epoch_loss / num_batches
    train_losses.append(avg_loss)
    
    # –í–∞–ª–∏–¥–∞—Ü–∏—è
    if epoch % 10 == 0 or epoch == num_epochs - 1:
        net.eval()
        with torch.no_grad():
            preds_valid = net(X_valid_t)
            rmse_epoch = float(torch.sqrt(loss_fn(preds_valid, y_valid_t)))
            val_rmse_list.append(rmse_epoch)
            
            if rmse_epoch < best_rmse:
                best_rmse = rmse_epoch
                best_epoch = epoch
                torch.save(net.state_dict(), 'best_model.pth')
            
            print(f'Epoch {epoch:3d}/{num_epochs} | '
                  f'Train Loss: {avg_loss:.4f} | '
                  f'Val RMSE: {rmse_epoch:.4f}')

display(f"–õ—É—á—à–µ–µ RMSE: {best_rmse:.4f}")
display(f"–õ—É—á—à–∞—è —ç–ø–æ—Ö–∞: {best_epoch}")
```

## –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏

–ü—Ä–æ–≤–µ–¥–µ–º —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏.

```python
def test_search_system(test_queries, test_image_folder, text_model, best_model_name):
    if not os.path.exists(test_image_folder):
        display(f"{test_image_folder}")
        return
    
    test_image_files = [f for f in os.listdir(test_image_folder) 
                       if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    display(f"{len(test_image_files)}")
    
    if not test_image_files:
        display("0")
        return
    
    test_image_paths = [os.path.join(test_image_folder, img) for img in test_image_files]
    test_img_features, successful_test_images = image_vectorizer.vectorize_images(test_image_paths)
    
    if len(test_img_features) == 0:
        display("0")
        return
    
    test_features_dict = dict(zip([os.path.basename(p) for p in successful_test_images], test_img_features))
    
    display(f"{len(test_features_dict)}")
    
    def search_images(query_text, features_dict, top_k=5):
        # –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è –∑–∞–ø—Ä–æ—Å–∞ —Å sentence-transformers
        query_embedding = text_model.encode([query_text])[0]
        
        # –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è
        text_normalizer = Normalizer(norm='l2')
        query_normalized = text_normalizer.transform([query_embedding])[0]
        
        scores = []
        
        for img_name, img_features in features_dict.items():
            image_normalizer = Normalizer(norm='l2')
            img_normalized = image_normalizer.transform([img_features])[0]
            
            # –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
            combined_features = np.concatenate([img_normalized, query_normalized])
            
            if best_model_name == "Linear Regression":
                relevance = model_lr.predict(combined_features.reshape(1, -1))[0]
            elif best_model_name == "Random Forest":
                relevance = rf_model.predict(combined_features.reshape(1, -1))[0]
            elif best_model_name == "Neural Network":
                with torch.no_grad():
                    relevance = net(torch.FloatTensor(combined_features).unsqueeze(0)).item()
            else:
                relevance = 0.0
            
            scores.append((img_name, relevance))
        
        scores.sort(key=lambda x: x[1], reverse=True)
        return scores[:top_k]
    
    test_samples = min(5, len(test_queries))
    
    display(f"{test_samples}")
    
    for i in range(test_samples):
        query_text = test_queries.iloc[i]['text']
        expected_image = test_queries.iloc[i]['image_name']
        
        display(f"{query_text}")
        display(f"{expected_image}")
        
        top_images = search_images(query_text, test_features_dict, top_k=5)
        
        for j, (img_name, score) in enumerate(top_images, 1):
            match_indicator = "‚úì" if img_name == expected_image else " "
            display(f"{j}. {match_indicator} {img_name} {score:.4f}")
        
        found_images = [img for img, _ in top_images]
        if expected_image in found_images:
            rank = found_images.index(expected_image) + 1
            display(f"{rank}")
        else:
            display("0")

# –ó–∞–ø—É—Å–∫ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è —Å –ª—É—á—à–µ–π –º–æ–¥–µ–ª—å—é
test_search_system(test_queries, TEST_IMAGES_PATH, text_model, best_model)
```
